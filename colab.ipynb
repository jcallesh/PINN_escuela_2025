# <span style="color:#1f77b4;">Google Colab</span>

Google Colab es un entorno de notebooks en la nube que permite ejecutar código Python sin necesidad de instalar nada en tu computadora.


## Consideraciones antes de empezar:

- Es ideal para aprender y experimentar.
- El tiempo de ejecución es limitado, especialmente en el caso de GPU.
- Puede desconectarse automáticamente si estás inactivo.
- Si utilizas demasiada memoria (RAM), el notebook puede reiniciarse y perderás todo lo que no esté guardado.
- Las sesiones en Google Colab no son permanentes y tampoco se guardan*.
- Si dejas de usar el notebook por un tiempo o ejecutas procesos muy largos*, Google puede cerrarlo automáticamente.


> Puedes usar una GPU con TensorFlow o PyTorch, pero el acceso no siempre está garantizado. Para comprobar si tu notebook tiene acceso a la GPU, ejecuta el siguiente comando en una celda:

```Python
    !nvidia-smi
```
Si una GPU está habilitada, retornara información sobre la tarjeta gráfica
